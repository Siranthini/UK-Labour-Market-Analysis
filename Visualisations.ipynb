```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
# To read the f
degree_df = pd.read_csv("employment_rates_by_degreeclass.csv")
disability_df = pd.read_csv("employment_rates_by_disability.csv")
ethinicity_df = pd.read_csv("employment_rates_by_ethnicity.csv")
gender_df = pd.read_csv("employment_rates_by_gender.csv")
```

```python
def to_percent(x, position):
    return f"{int(x)}%"
```

```python

degree_df = degree_df[degree_df['graduate_group'] == 'Graduates 16-64']

# To Melt the dataframe to seaborn format
melted_df = degree_df.melt(id_vars=['Degree_class'], 
                              value_vars=['Employment_rate', 'High_skill_emp_rate', 'Unemployment_rate', 'Inactivity_rate'],
                              var_name='Rate_type', value_name='Value')

# To set the colour palatte to differentiate Degree Class
palette = sns.color_palette('husl', n_colors=len(melted_df['Degree_class'].unique()))

# To create a bar plot
plt.figure(figsize=(15, 8))
barplot = sns.barplot(x='Value', y='Rate_type', hue='Degree_class', data=melted_df, palette='YlOrBr')

# To format the values displayed in X axis
formatter = FuncFormatter(to_percent)
plt.gca().xaxis.set_major_formatter(formatter)

plt.xticks(rotation=45)
plt.title('Graduate Employment Rates by Degree Class: Working Age group (16-64 years), 2022')
plt.xlabel('Employment Rate (%)')
plt.ylabel('')
plt.legend(title='Degree Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
